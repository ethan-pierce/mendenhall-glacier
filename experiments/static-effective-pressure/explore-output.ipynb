```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 4, figsize = (20, 5))

Ns = [60, 80, 90, 95]
shape = (224, 197)

disp = []

for i in range(4):
    dispersed = np.loadtxt('./outputs/slow/spatial/disp_Pw_' + str(Ns[i]) +'.txt')
    disp.append(dispersed)
    
    im = ax[i].imshow(np.flip(np.reshape(dispersed, shape), axis = 0), cmap = 'jet', vmax = 2.0)
    cbar = plt.colorbar(im, ax = ax[i], fraction = 0.0543, pad = 0.04)

plt.tight_layout()
plt.show()
```
